```python
import os 
import cv2 
from PIL import Image 
import pytesseract as pt 
```

```python
#defining the show_image function the pops up the image dialog box
#this image dialog box is actually the image that is being fed up
#to the system.

def show_image(img_path, size=(500, 500)):
    image = cv2.imread(img_path)
    image = cv2.resize(image, size)
    
    cv2.imshow("IMAGE", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
#defining the path of the module is necessary for the system to 
#run. This path is being set up to the env variables in the running
#os of the system. Though, this is an optional step and does not 
#necessarily need to set up the path. But, to avoid the ambiguity in the
#system, we set up the path. The given path is the 'ABSOLUTE PATH' for the 
#tesseract OCR engine.

pt.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
#Extracting the text from the images and the time-out exception.
#Timeout happens when the model is not able to recognise
#the characters, that is not able to detect the characters
#in the test image.

#defining the path of the test image.
path = create_path("33-other-empty.png")

image = Image.open(path)

#catching the exception in case no characters are found in
#the test image, followed by returning the time-out error message.
try:
    text = pt.image_to_string(image, lang='eng', timeout=2)

except RuntimeError as timeout_error:
    print("[TIMEOUT ERROR]")
    text = ''

# Check if the image has no detected text
if text == '':
    text = 'NO TEXT TO BE APPEARED'

print(text)
show_image(path)
image.close()
```

```python
path = create_path("34-other-news-template.png")

image = Image.open(path)
bound_rects = pt.image_to_boxes(image)

print(bound_rects)
show_image(path)
```

```python
#assigning the box bounds to their positions based on the 
#previous code block.

img = cv2.imread(path)
h, _, _ = img.shape

for b in bound_rects.splitlines():
    b = b.strip().split(' ')
    img = cv2.rectangle(img, (int(b[1]), h - int(b[2])), (int(b[3]), h - int(b[4])), (0, 255, 0), 2)

cv2.namedWindow("CHARACTERIZED IMAGE", cv2.WINDOW_NORMAL)
cv2.resizeWindow("CHARACTERIZED IMAGE", 400, 400)

cv2.imshow("CHARACTERIZED IMAGE", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Get verbose data including boxes, confidences, line and page numbers

#This function selects an image file from a predefined list and constructs its full path. It
#then loads the image using the Pillow library. Next, it applies Optical Character Recognition
#(OCR) using PyTesseract to extract text and metadata from the image. The extracted text is 
# printed,and finally, the image is displayed.

image_path = test_image_files[33]
path = create_path(image_path)

image = Image.open(path)
text = pt.image_to_data(image)

print(text)
show_image(path)

#time.sleep(60)
#image.close()
```
